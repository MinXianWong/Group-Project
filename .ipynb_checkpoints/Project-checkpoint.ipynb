# Introduction

## Background Information

How do plants grow? Plants take in water and nutrients from the soil through their roots, but most of the water in the plant is lost to the surroundings via transpiration. A small amount of the remaining water is used by the leaves for photosynthesis to produce biomass for the plant's growth. A plant's growth is dependent on many factors, such as the amount of water it has, its intrinsic rate of growth (how fast it grows naturally - there are fast-growing and slow-growing plants), and the rate of plant degradation (the plant uses the biomass it produces for respiration). To determine how much a plant has grown, we can measure the dry mass of the plant (the dry weight of a plant, excluding all of the water in it). 

So, how does the amount of water in plants affect its growth? 

Photosynthesis: $6CO_2 + 6H_2O$ &rarr; $C_6H_{12}O_6 + 6O_2$ 

As seen from this equation, the production of each molecule of glucose (biomass) requires 6 molecules of water. Thus, a plant will grow well with an optimal amount of water, but is unable to grow and will die if it has no water.

As such, the amount of water available in the soil for the plant to take up is important as well. The amount of water in the soil is dependent on the amount of input - precipitation and irrigation, and the amount of output - evaporation and transpiration. When the amount of water in the soil is low, there is less water available for plants to take up, which slows plant growth. However, when there is too much water in the soil, the roots cannot get sufficient oxygen and will die, causing the plant to die as well. Hence, for plant growth to be optimal, the amount of water in the soil has to be monitored and well-regulated too. 


## Project Direction and Scope

In this project, we aim to model the effect of irrigation on plant growth. We will be focusing mainly on the presence or absence and magnitude of irrigation, which affects the amount of water in the soil and plant, ultimately affecting the growth of the plant. Factors such as evaporation, transpiration and plant growth characteristics will also be considered. For this model, other environmental factors such as light, nutrients, temperature and air quality will be assumed as ideal and has no influence on plant growth and water conditions. While the model would be more useful if other environmental factors were considered as well, it would be extremely complex to model as there are too many factors that are interlinked. 

Our model will be useful for intensive farming practices, where environmental factors are highly controlled as compared to traditional farming. Intensive farming is becoming more prevalent globally in order to meet increasing food demands by the growing world population. To supplement farmers' efforts in optimising crop growing conditions, this model can find optimal irrigation techniques and maximize their crop yields.


## Project Significance

Maximizing crop yields is extremely important as the increasing demand becomes harder to meet. Crop yields are decreasing due to poorly-managed soils, extreme weather, and pollution from human activities, and there is a limit to the use of fertilizers to increase crop yields. Furthermore, we are placing a lot of stress on our water resources, so it is crucial that we pay attention to our water consumption in agriculture - especially since it is one of the most water-intensive industries in the world.  

A whooping [60 percent of water used in agriculture is wasted](https://earth.org/smart-farms-how-ai-can-solve-water-wastage), but we can [save water through optimizing watering times](https://www.sigfox.com/8-tips-agricultural-water-conservation/) [according to soil moisture](https://greentumble.com/ways-to-reduce-water-consumption-on-farms). Other forms of agriculture like hydroponics [can help reduce water usage too](https://www.nps.gov/articles/hydroponics.htm), and our model will be very relevant to water-based agriculture.

This model can also predict plant growth under various conditions, allowing more food to be [produced more efficiently](https://www.frontiersin.org/articles/10.3389/fpls.2022.1069190), [more effective conservation efforts for endangered species of plants](http://www.jstor.org/stable/45220869), and understanding [how plants respond to climate change](https://www.mdpi.com/2073-4441/14/17/2728). In fact, water is known to be [the most limiting abiotic factor in determining plant growth](https://www.nature.com/scitable/knowledge/library/water-uptake-and-transport-in-vascular-plants-103016037/)

As a result, this study ultimately contributes to conservation and sustainable agricultural production.


## Model of Interest

Using [this paper](https://www.mdpi.com/2227-7390/10/1/151) as a reference, we used several parameters and differential equations to model the impact of water availability on plant growth rate. 

The 3 differential equations in the paper represent the relationship between plant dry mass, mass of water in the plant, and mass of water in the soil aroudn the plant over time.

<pre>

</pre>

Rate of increase in plant dry mass:
\begin{equation}
x'(t) = \dfrac{k_1 x(t)w(t)}{1+gx(t)} - mx(t)
\newline
\hspace{30pt} \downarrow \hspace{35pt} \downarrow
\newline
\hspace{30pt} (A) \hspace{27pt} (B)
\end{equation}

(A) $\rightarrow$ Increase in dry mass due to photosynthesis

(B) $\rightarrow$ Basal rate of degradation

<pre>


</pre>

Changes in water content in plant:
\begin{equation}
w'(t) = d\dfrac{w(t)v(t)}{1+rw(t)} - b w(t) - \dfrac {k_2x(t)w(t)}{1+gx(t)}
\newline
\hspace{29pt} \downarrow \hspace{35pt} \downarrow \hspace{35pt} \downarrow
\newline
\hspace{30pt} (C) \hspace{27pt} (D) \hspace{27pt} (E)
\end{equation}

(C) $\rightarrow$ Water taken up by plant

(D) $\rightarrow$ Transpiration & other losses

(E) $\rightarrow$ Water used for photosynthesis  

<pre>


</pre>

Changes in soil water availability:
\begin{equation}
v'(t) = -y v(t) - p\dfrac{w(t)v(t)}{1+rw(t)}
\newline
\hspace{20pt} \downarrow \hspace{37pt} \downarrow
\newline
\hspace{23pt} (F) \hspace{29pt} (G)
\end{equation}

(F) $\rightarrow$ Evaporation & other losses

(G) $\rightarrow$ Water taken up by plant


where $x(t)$, $w(t)$ and $v(t)$ represents the plant drymass/size, water content in plant available for growth and soil water availability at time $t$ respectively.

The parameters used are:


|Parameters| Meaning | Default Values | Units |
|:--:| :--: | :---:  | :--: |
| $k_1$ |Intrinsic growth rate per unit of water inside the plant | 0.01 | \[time x mass\]$^{-1}$ |
| $g$   | Limiting factor constant of $x(t)$                      | 0.1  |    \[mass\]$^{-1}$     |
| $m$   |   Plant degradation rate                                | 0.00001|    \[time\]$^{-1}$    |
| $d$ | Intrinsic rate of increase of the water inside the plant | 1 | \[time x mass\]$^{-1}$ |
| $r$ |   Limiting factor constant of $w(t)$                 |  20 |   \[mass\]$^{-1}$    |
| $b$ |   Rate of decrease of water inside the plant           | 0.1      |     \[time\]$^{-1}$    |
| $k_2$ | Intrinsic rate of water decrease by photosynthesis   | 0.01  | \[time x mass\]$^{-1}$ |
| $y$  |   Inner rate of decrease of the pond water           | 0.000009 |    \[time\]$^{-1}$    |
| $p$ | Intrinsic rate of water that goes to the plant      | 2   | \[time x mass\]$^{-1}$ |

The default values for these parameters were largely arbitrarily determined for the model to make sense numerically.

# Implementing the Model in Python

## Model Direct Implementation

We are first generating a few helper functions so that our model implementation will be easier to parse. These helper functions aren't super important and you can skip them for now.

But if you want to know, TLDR of our 3 main helper functions are below.

`generate_hold_lists()` creates 4 lists for storing all values of timepoints, x, w, and v as our model runs.

`update_next_timepoint()` updates our generated lists and is in charge of carrying out the calculations at every timepoint.

`draw_nice_graph()` takes in our final produced lists of values and plots it nicely to our specifications.

```python
def generate_hold_lists(num_ticks, x_0, w_0, v_0):
    '''
    This function takes in 4 integer inputs.
    num_ticks: the number of ticks we want our simulation to run for
    x_0, w_0, v_0: the values of x, w, and v respectively at time 0
    
    The function outputs a tuple containing 4 lists, which hold all the timepoints, x values, w values, and
    v values respectively.
    '''
    all_timepoints = [timepoint for timepoint in range(0, num_ticks+1)]
    x_list = [x_0]
    w_list = [w_0]
    v_list = [v_0]
    
    return (all_timepoints, x_list, w_list, v_list)
```

```python
def update_next_timepoint(parameter_values, x_list, w_list, v_list):
    '''
    Takes in 4 iterables, parameter_values, x_list, w_list, v_list.
    parameter_values has to be in the sequence (k_1, g, m, d, r, b, k_2, y, p).
    These input parameters provide the necessary information for this function to update the lists referenced.
    Since lists are mutable, this function does not return any value but directly manipulates the lists passed.
    '''
    k_1, g, m, d, r, b, k_2, y, p = parameter_values        # unpack the input iterables 
    
    last_x = x_list[-1]                    # find the x, w, v at the last timepoint
    last_w = w_list[-1]
    last_v = v_list[-1]
    
    # calculate changes in x, w, and v according to the given formulae
    
    x_change = (k_1 * last_x * last_w/(1 + g * last_x)) - (m * last_x)
    w_change = (d * last_w * last_v / (1 + r * last_w)) - (b * last_w) - (k_2 * last_x * last_w / (1 + g * last_x))
    v_change = (-y * last_v) - (p * last_w * last_v / (1 + r * last_w))
    
    new_x = last_x + x_change
    new_w = last_w + w_change
    new_v = last_v + v_change
    
    x_list.append(new_x)
    w_list.append(new_w)
    v_list.append(new_v)
```

```python
def create_graph_info(all_timepoints, plot_1, plot_2, plot_3):
    '''
    Takes in 4 parameters: all_timepoints, plot_1, plot_2, and plot_3.
    all_timepoints is a list of all timepoints to serve as the x-axis.
    plot_1, plot_2, and plot_3 are the plotting information for plots 1, 2, and 3 respectively which are tuples
    that contain a string for the plot label and the list of the raw data to be plotted.
    The function outputs a single tuple that contains all these parameters.
    '''
    all_plot_info = (all_timepoints, ) + (plot_1, ) + (plot_2, ) + (plot_3, )
    return all_plot_info     # returns a tuple first with the list of timepoints followed by the plot info of every plot

# ^ each one of these will be termed as "graphs" or "graph_info"
```

```python
def draw_nice_graph(*graphs):
    '''
    Takes in a variable number of graph plot infos anywhere from 1 to 4 plots and creates nicely spaced out graphs. 
    '''
    
    colors = ['tab:green', 'tab:blue', 'tab:orange']   # colors used for each line in each graph
    
    if len(graphs) == 1:     # graphs without the asterisk will place all included arguments in a tuple
        fig, ax = plt.subplots(ncols=1, nrows=1, figsize=(5, 5))
        
        for graph_data in graphs:
            for i, y_values in enumerate(graph_data[1:]):    # for each plotted line for each graph
                ax.plot(graph_data[0], y_values[1], label=y_values[0], color=colors[i])     # plot all_timepoints on x-axis, 
                                                                           # and y-values for each plotted line on y-axis
            ax.legend()
            ax.set_xlabel("Time / Arbitrary Units")
            ax.set_ylabel("Value / Arbitrary Units")
    
    elif len(graphs) < 4:
        fig, ax = plt.subplots(ncols=len(graphs), nrows=1, figsize=(12, 5))
        
        for a, graph_data in enumerate(graphs):
            for i, y_values in enumerate(graph_data[1:]):
                ax[a].plot(graph_data[0], y_values[1], label=y_values[0], color=colors[i])  # plots with multiple columns but only 1 row
                                                                           # have ax as a one-layer deep nested list
            ax[a].legend()
            ax[a].set_xlabel("Time / Arbitrary Units")
            ax[a].set_ylabel("Value / Arbitrary Units")
    
    else:
        fig, ax = plt.subplots(ncols=2, nrows=2, figsize=(10, 8))
        
        for j, graph_data in enumerate(graphs):
            a = 1 if j > 1 else 0
            b = 1 if j % 2 else 0
            
            for i, y_values in enumerate(graph_data[1:]):
                ax[a][b].plot(graph_data[0], y_values[1], label=y_values[0], color=colors[i])
            
            ax[a][b].legend()
            ax[a][b].set_xlabel("Time / Arbitrary Units")
            ax[a][b].set_ylabel("Value / Arbitrary Units")
    
    plt.show()
```

Now that we have constructed the necessary helper functions, we can move on to our main model and the implementation of it! The model is quite simple in terms of code.

```python
def model_plant_growth(x_0, w_0, v_0, time_ticks = 300):
    '''
    A function that takes in initial values of x(0), w(0), and v(0).
    Also takes in an optional argument time_ticks, an integer representing the number of ticks to run the program for,
    which is 300 by default.
    Output is a plot of x(t), w(t), and v(t) over time.
    '''
    
    ### Default Variables --- Can be changed! ###
    all_params = (0.01, 0.1, 0.00001, 1, 20, 0.1, 0.01, 0.000009, 2)               
    
    ### Generating our lists to hold all relevant values ###
    all_timepoints, x_list, w_list, v_list = generate_hold_lists(time_ticks, x_0, w_0, v_0)

    ### Filling up our lists of values ###
    for timepoint in all_timepoints[1:]:          
        update_next_timepoint(all_params, x_list, w_list, v_list)
    
    ### Creating tuples with labels for each list ###
    x_info = ("Plant Dry Mass", x_list)
    w_info = ("Mass of water in plant", w_list)
    v_info = ("Mass of water in soil", v_list)
    
    return create_graph_info(all_timepoints, x_info, w_info, v_info)
```

## Model Output Verification

Now that we have defined our model as a function, and this function generates a plot, we can try running an instance of it to check that it works!

```python
graph = model_plant_growth(3,7,20, time_ticks = 150)
draw_nice_graph(graph)
```

Looking at this graph, our model seems to generally produce expected results. The mass of water in the soil steadily decreases to 0, as does the mass of water in the plant. Subsequently, plant dry mass can no longer increase once the mass of water in the plant reaches 0.

## Model Extension (Optional Arguments)

The obvious manner in which we could provide irrigation to our plant is by modifying our `update_next_timepoint()` function. If this function were able to take in a Boolean about whether or not to irrigate a plant, as well as information on the rules for irrigation (in terms of quantity and when to irrigate), then we could update `change_v` accordingly by adding our irrigation quantity to `change_v`.

Let's name this new updated `update_next_timepoint()` function as `update_next_timepoint_v1()`. It is very similar to `update_next_timepoint()`, as it contains all of its functionalities. The major difference is that it takes in an additional 3 parameters, and the docstring under this function explains how these 3 parameters are used.

Before we tackle the implementation of irrigation to our model, we have to first consider the specifications of this irrigation process. In the paper that we extracted this model from, plant irrigation was done at regular time intervals. However, determining the time intervals at which to irrigate plants is difficult: irrigation may be done too regularly, such that the amount of water in the soil is increasing at a faster rate than it is decreasing. Alternatively, irrigation may be too infrequent, such that the amount of water in the soil is reaching 0 for extended periods of time before irrigation occurs again.

Hence, we propose a different rationale behind irrigation rules: we will irrigate the plant whenever the amount of water in the soil falls below a certain threshold. This more closely emulates [the watering processes used in agriculture](https://cals.arizona.edu/extension/ornamentalhort/waterquality/irrigation_when.pdf), and is hence a more useful way to build our model. To accommodate the specification of a threshold and watering quantity, our function `update_next_timepoint_v1()` will take in `irrigate_threshold` and `irrigate_qty` to specify these rules.

```python
def update_next_timepoint_v1(parameter_values, x_list, w_list, v_list, 
                             irrigate = False, irrigate_threshold = 5, irrigate_qty = 10):
    '''
    Updated in v1 to take in 3 additional parameters: irrigate, irrigate_threshold, irrigate_qty.
    irrigate is a boolean that defaults to False.
    irrigate_threshold is an integer that defaults to 5, representing the threshold of v(t) below which we will water.
    irrigate_qty is an integer that defaults to 10, representing the quantity of water to input into our system 
    whenever we do irrigate it. This quantity will be added to new_v to simulate pouring water onto the soil around a plant.
    '''
    k_1, g, m, d, r, b, k_2, y, p = parameter_values        # unpack the input iterables 
    
    last_x = x_list[-1]                    # find the x, w, v at the last timepoint
    last_w = w_list[-1]
    last_v = v_list[-1]
    
    x_change = (k_1 * last_x * last_w/(1 + g * last_x)) - (m * last_x)         # calculate changes in x, w, and v
    w_change = (d * last_w * last_v / (1 + r * last_w)) - (b * last_w) - (k_2 * last_x * last_w / (1 + g * last_x))
    v_change = (-y * last_v) - (p * last_w * last_v / (1 + r * last_w))
    
    new_x = last_x + x_change
    new_w = last_w + w_change
    new_v = last_v + v_change
    
    # 2 conditions for irrigation: we are told to, and soil water level is below irrigate_threshold
    if irrigate and new_v < irrigate_threshold:              
        new_v += irrigate_qty
    
    x_list.append(new_x)
    w_list.append(new_w)
    v_list.append(new_v)
```

Now that we have successfully defined an updated function `update_next_timepoint_v1()` that can account for irrigation, we can produce a second `model_plant_growth()` that can handle this irrigation. We will call this `model_plant_growth_v1()` for consistency.

To allow our `model_plant_growth_v1()` to pass correct parameters to our `update_next_timepoint_v1()`, our new function also has to take in these 3 parameters to pass them to our helper function `update_next_timepoint_v1()`.

```python
def model_plant_growth_v1(x_0, w_0, v_0, time_ticks = 300, 
                          irrigate = False, irrigate_threshold = 5, irrigate_qty = 10):
    '''
    Updated in v1 to take in 3 new inputs irrigate, irrigate_threshold, and irrigate_qty.
    These parameters are then passed to the function that carries out calculations of x(t), w(t) and v(t).
    '''
    
    ### Default Variables --- Can be changed! ###
    all_params = (0.01, 0.1, 0.00001, 1, 20, 0.1, 0.01, 0.000009, 2)              
    
    ### Generating our lists to hold all relevant values ###
    all_timepoints, x_list, w_list, v_list = generate_hold_lists(time_ticks, x_0, w_0, v_0)

    ### Filling up our lists of values ###
    for timepoint in all_timepoints[1:]:          
        update_next_timepoint_v1(all_params, x_list, w_list, v_list, irrigate, irrigate_threshold, irrigate_qty)
    
    ### Creating tuples with labels for each list ###
    x_info = ("Plant Dry Mass", x_list)
    w_info = ("Mass of water in plant", w_list)
    v_info = ("Mass of water in soil", v_list)
    
    return create_graph_info(all_timepoints, x_info, w_info, v_info)
```

Now that we have a new function `model_plant_growth_v1()`, we can attempt to plot the same parameters as before and observe what happens if we set irrigate to `True`!

```python
graph = model_plant_growth_v1(3,7,20, irrigate = True, time_ticks = 150)
draw_nice_graph(graph)
```

# Improvements to Model

When we try to play around with the model a little more, we start to observe that there are some peculiarities with this model, especially if we change the default parameters. Since many of these default parameters are arbitrarily manipulated to match with empirical data, and are not grounded in scientific basis, we can consider ways to improve this model for a greater range of default parameter values.

At the present moment, our model has its default parameters defined as a variable inside a function, and as such we cannot directly change these parameters without editing the function itself. We can begin by improving our model function such that it is able to take in these parameters. This will be `model_plant_growth_v2()`. For easy reference to our previous default parameter values, we can create a global variable `default_params` such that we can choose to use these values in our plotting.

```python
default_params = (0.01, 0.1, 0.00001, 1, 20, 0.1, 0.01, 0.000009, 2)
```

## Increasing Model Customisability

```python
def model_plant_growth_v2(x_0, w_0, v_0, parameters, time_ticks = 300, 
                          irrigate = False, irrigate_threshold = 5, irrigate_qty = 10):
    '''
    Updated in v2 to take in a new parameter, named: parameters. This will be
    an iterable in the sequence k_1, g, m, d, r, b, k_2, y, p. This allows for greater customisability
    in our model, making it easier to change these parameters.
    '''           
    
    ### Generating our lists to hold all relevant values ###
    all_timepoints, x_list, w_list, v_list = generate_hold_lists(time_ticks, x_0, w_0, v_0)

    ### Filling up our lists of values ###
    for timepoint in all_timepoints[1:]:          
        update_next_timepoint_v1(parameters, x_list, w_list, v_list, irrigate, irrigate_threshold, irrigate_qty)
    
    ### Creating tuples with labels for each list ###
    x_info = ("Plant Dry Mass", x_list)
    w_info = ("Mass of water in plant", w_list)
    v_info = ("Mass of water in soil", v_list)
    
    return create_graph_info(all_timepoints, x_info, w_info, v_info)
```

As demonstrated by the 2 plots above, our model is able to produce the same plots as previously, as well as produce new results depending on changes to the parameters passed to the function! Hence, our new function is more customizable with modelling using different parameters now.

## Account for Negative Values

One problem with our model right now is that some parameter values can result in a graph that appears to not make sense. Consider the instance below where we use parameters identical to the default_params, but changing `r`, the limiting factor constant of $w(t)$ from 20 to 1:

```python
problem_params = (0.01, 0.1, 0.00001, 1, 1, 0.1, 0.01, 0.000009, 2)   # same as default_params, but r is changed from 20 to 1 
problem_graph = model_plant_growth_v2(3,7,20, problem_params, time_ticks = 50)
draw_nice_graph(problem_graph)
```

In this example, the lack of checking for negative v(t) values in the presence of a high $r$ causes w(t) to oscillate wildly between positive and negative values. To change this, we could modify our model with simple if-else conditions. We know in reality that dry mass of a plant, mass of water in a plant, and mass of water in the soil will never be negative. As such, we simply incorporate new if-else conditionals that set x(t), w(t), or v(t) to 0 when they are negative.

Below is an updated function that calculates the next timepoint but checks for negative values of $x(t)$, $w(t)$ and $v(t)$.

```python
def update_next_timepoint_v2(parameter_values, x_list, w_list, v_list, 
                             irrigate = False, irrigate_threshold = 5, irrigate_qty = 10):
    '''
    Updated in v2 to check for negative values of x, w, and v.
    If these values are negative, they will be set to 0 instead.
    '''
    k_1, g, m, d, r, b, k_2, y, p = parameter_values        # unpack the input iterables 
    
    last_x = x_list[-1]                    # find the x, w, v at the last timepoint
    last_w = w_list[-1]
    last_v = v_list[-1]
    
    x_change = (k_1 * last_x * last_w/(1 + g * last_x)) - (m * last_x)         # calculate changes in x, w, and v
    w_change = (d * last_w * last_v / (1 + r * last_w)) - (b * last_w) - (k_2 * last_x * last_w / (1 + g * last_x))
    v_change = (-y * last_v) - (p * last_w * last_v / (1 + r * last_w))
    
    new_x = max(last_x + x_change, 0)         # if new_x, new_w or new_v are < 0, they are set to 0 instead
    new_w = max(last_w + w_change, 0)
    new_v = max(last_v + v_change, 0)
    
    # 2 conditions for irrigation: we are told to, and soil water level is below irrigate_threshold
    if irrigate and new_v < irrigate_threshold:              
        new_v += irrigate_qty
    
    x_list.append(new_x)
    w_list.append(new_w)
    v_list.append(new_v)
```

```python
def model_plant_growth_v3(x_0, w_0, v_0, parameters, time_ticks = 300, 
                          irrigate = False, irrigate_threshold = 5, irrigate_qty = 10):
    '''
    Updated in v3 to call the updated function update_next_timepoint_v2.
    '''          
    
    ### Generating our lists to hold all relevant values ###
    all_timepoints, x_list, w_list, v_list = generate_hold_lists(time_ticks, x_0, w_0, v_0)

    ### Filling up our lists of values ###
    for timepoint in all_timepoints[1:]:          
        update_next_timepoint_v2(parameters, x_list, w_list, v_list, irrigate, irrigate_threshold, irrigate_qty)
    
    ### Creating tuples with labels for each list ###
    x_info = ("Plant Dry Mass", x_list)
    w_info = ("Mass of water in plant", w_list)
    v_info = ("Mass of water in soil", v_list)
    
    return create_graph_info(all_timepoints, x_info, w_info, v_info)
```

```python
problem_params = (0.01, 0.1, 0.00001, 1, 1, 0.1, 0.01, 0.000009, 2)   # same as default_params, but r is changed from 20 to 1 
problem_graph = model_plant_growth_v3(3,7,20, problem_params, time_ticks = 50)

draw_nice_graph(problem_graph)
```

With the addition of these conditionals, we now see that our graph makes much more sense. `r` is a constant that limits the rate at which our plant can take up water, so changing the value of `r` from 20 to 1 is not impossible. This simply means a lower restriction on how fast our plant can take up water! Lowering our `r` value from 20 to 1 means a lower limit on rate of plant water uptake, and hence means much faster water uptake by our plant. As a result, we obtain the graph above: where the mass of water in plant shoots up sharply and the mass of water in soil drops very quickly, since our hypothetical plant is absorbing water at a ludicrous rate. Of course, this value of `r` and rate of water uptake by a plant is impractical, but this example simply illustrates the importance of checking for negative values to avoid such edge cases.

The previous model was unable to handle this as it did not check for negative values, but our improved `model_plant_growth_v3()` does. Now we can play around with our model with the assurance that it won't create negative numbers where it shouldn't!

## Improvement on default parameters

Recall that some of our parameters, such as r and p, were arbitrarily manipulated in our paper to fit the curves. As such, many of the default proposed values can be subject to change to better model certain aspects of plant growth. There are a few changes we propose to these parameters and their default values, as listed below:

1. `d` and `p` are two variables that are needlessly distinguished, and can be merged into the new variable `q`
    - `d` is a constant representing the intrinsic rate of increase of water in the plant (per unit time per unit mass).
    - `p` is a constant representing the intrinsic rate of water that goes from the pond to the plant (per unit time per unit mass).
    - When we examine our equations for $w'(t)$ and $v'(t)$, we can observe that both the values of $w'(t)$ and $v'(t)$ are affected by $\frac{w (t) v (t)}{1 + r w (t)}$. $w'(t)$ is affected by positive $d \frac{w (t) v (t)}{1 + r w (t)}$, whereas $v'(t)$ is affected by negative $p\frac{w (t) v (t)}{1 + r w (t)}$. 
    - Since d and p were arbitrarily determined in the first place, we can try out an arbitrary value of the new variable q that is just the mean of their two default values: q: $\frac{2+1}{2} = 1.5$
    
    
2. k_1 and k_2 can be represented by the same variable `k`
    - `k_1` is the intrinsic growth rate per unit of water in the plant (per unit time per unit mass)
    - `k_2` is the intrinsic rate of water decrease by photosynthesis (per unit time per unit mass)
    - Therefore, `k_1` and `k_2` should be closely related to each other.
    - In the paper that our model was adapted from, `k_1` and `k_2` are assumed to be approximately constant and are hence assigned the same default value.
    - By the photosynthetic equation mentioned earlier, $6 CO_2 + 6 H_2O \rightarrow C_6H_{12}O_6 + 6 O_2$, it can be calculated that 108g of $H_2O$ is needed to produce 180g of $C_6H_{12}O_6$, which reduces into the ratio $3:5$. This tells us that $k_1 = \frac{5}{3} k_2$
    - Hence, we only need a single input `k`, and calculations can then be carried out within our function to break down the parameter `k` into internal `k_1` and `k_2` variables. For convention, we can have k = k_1, such that k_2 = $\frac{5}{3}$`k`
    - This simplifies the parameters input for our function, reducing unnecessary complexity. It also provides a scientific basis upon which these values are determined, instead of arbitrary assignment for model fit.
   
   
3. Default `m` value can be increased to 0.01.
    - `m` is the degradation rate of our plant dry mass (per unit time)
    - Default `m` value in the paper is 0.00001, but this value is far too low for any serious plant degradation to be observed, even if the plant goes without water for very long timepoints. If we change the m value by increasing it, we start to notice non-negligible degradation of our plant dry mass.
    - Furthermore, this sets an upper limit on how high $x(t)$ can go, modelling the maximal plant growth.
    - For different plants, the value of m would be different (depending on their resilience to harsh conditions / lifespan), but this default value of 0.01 is more representative of most herbaceous plants (which form the bulk of agricultural crops) that  can't last too long without water, and do not live as long as woody plants.
    - For different plants with different maximum growth capabilities, this value of `m` can be edited under parameters to model that specific plant's maximum dry mass.
    
    
To implement this, we will redefine our default_params to match these updated values.

```python
default_params = (0.01, 0.1, 0.01, 1.5, 20, 0.1, 0.000009)
```

```python
def update_next_timepoint_v3(parameter_values, x_list, w_list, v_list, 
                             irrigate = False, irrigate_threshold = 5, irrigate_qty = 10):
    '''
    Updated in v3 to handle new structure for parameters.
    d and p are merged into q.
    k_1 and k_2 are defined in this function based off the input k.
    '''
    k, g, m, q, r, b, y = parameter_values        # unpack the input iterables 
    k_1, k_2 = k, k*5/3                           # assign k_1 and k_2
    
    last_x = x_list[-1]                    # find the x, w, v at the last timepoint
    last_w = w_list[-1]
    last_v = v_list[-1]
    
    x_change = (k_1 * last_x * last_w/(1 + g * last_x)) - (m * last_x)         # calculate changes in x, w, and v
    w_change = (q * last_w * last_v / (1 + r * last_w)) - (b * last_w) - (k_2 * last_x * last_w / (1 + g * last_x))
    v_change = (-y * last_v) - (q * last_w * last_v / (1 + r * last_w))
    
    new_x = max(last_x + x_change, 0)         # if new_x, new_w or new_v are < 0, they are set to 0 instead
    new_w = max(last_w + w_change, 0)
    new_v = max(last_v + v_change, 0)
    
    # 2 conditions for irrigation: we are told to, and soil water level is below irrigate_threshold
    if irrigate and new_v < irrigate_threshold:              
        new_v += irrigate_qty
    
    x_list.append(new_x)
    w_list.append(new_w)
    v_list.append(new_v)
```

```python
def model_plant_growth_v4(x_0, w_0, v_0, parameters, time_ticks = 300, 
                          irrigate = False, irrigate_threshold = 5, irrigate_qty = 10):
    '''
    Updated in v4 to combine variables p and d. Variables k_1 and k_2 are also combined into k.
    default_params is also overridden concurrently in defining this function to match these changes.
    Value of parameter m in default_params is also updated from 0.00009 to 0.01
    Also calls the new update_next_timepoint_v3()
    '''            
    
    ### Generating our lists to hold all relevant values ###
    all_timepoints, x_list, w_list, v_list = generate_hold_lists(time_ticks, x_0, w_0, v_0)

    ### Filling up our lists of values ###
    for timepoint in all_timepoints[1:]:          
        update_next_timepoint_v3(parameters, x_list, w_list, v_list, irrigate, irrigate_threshold, irrigate_qty)
    
    ### Creating tuples with labels for each list ###
    x_info = ("Plant Dry Mass", x_list)
    w_info = ("Mass of water in plant", w_list)
    v_info = ("Mass of water in soil", v_list)
    
    return create_graph_info(all_timepoints, x_info, w_info, v_info)
```

## Account for overwatering

One factor that is inadequately simulated by our current model is modelling of over-watering. When our plant goes for an extended period of time with insufficient water, our new `m` value results in plant degradation. This models the occurrence of plant death when insufficient water is provided.

However, over-watering of plants is not modelled by our current model. To solve this, we could edit our `update_next_timepoint_v4` and `model_plant_growth_v4` to take in an additional input, drown_threshold. When $v(t)$ is at or below drown_threshold, the value of `m` (the plant degradation rate) will decrease multiplicatively. And when $v(t)$ is above `drown_threshold`, the value of `m` will recover up to its initial rate. This allows us to model the plant dying as it continues to be overwatered, as well as its recovery from overwatering when irrigation quantity becomes suitable.

```python
 def model_plant_growth_v5(x_0, w_0, v_0, parameters, time_ticks = 300,
                          irrigate = False, irrigate_threshold = 5, irrigate_qty = 10, drown_threshold = 30):
    '''
    Updated in v5 to account for plant degradation rate change when overwatered.
    Takes in an additional input drown_threshold, with default value 30.
    Typecasts parameters to a list so that the new update_next_timepoint_v4 can edit m as necessary.
    '''            
    
    parameters = list(parameters)                       # typecasting parameters to a list for in-place modification
    initial_m = parameters[2]                           # hold initial starting m value
    
    ### Generating our lists to hold all relevant values ###
    all_timepoints, x_list, w_list, v_list = generate_hold_lists(time_ticks, x_0, w_0, v_0)

    ### Filling up our lists of values ###
    for timepoint in all_timepoints[1:]:    
        if v_list[-1] >= drown_threshold:                 # if the most recent value of v >= drown_threshold,
            parameters[2] *= 4                            # the m value gets multiplied by 2
        elif parameters[2] > initial_m:                                            # if the most recent value of v < drown_threshold,
            parameters[2] /= 2
        update_next_timepoint_v3(parameters, x_list, w_list, v_list, irrigate, irrigate_threshold, irrigate_qty)
    
    ### Creating tuples with labels for each list ###
    x_info = ("Plant Dry Mass", x_list)
    w_info = ("Mass of water in plant", w_list)
    v_info = ("Mass of water in soil", v_list)
    
    return create_graph_info(all_timepoints, x_info, w_info, v_info)
```

Now that we have defined a new `model_plant_growth_v5()` that can account for overwatering of plants, we can compare our old and new graphs to evaluate whether this new function accurately models what we would expect to see in a plant being overwatered.

```python
old_overwatered = model_plant_growth_v4(3,7,20, default_params, irrigate = True, irrigate_qty = 35)
new_overwatered = model_plant_growth_v5(3,7,20, default_params, irrigate = True, irrigate_qty = 35)

draw_nice_graph(old_overwatered, new_overwatered)
```

The above graphs are of the older and newer model, respectively. In the old model, overwatering of a plant simply results in its dry mass increasing to a great extent. However, the new model here dies after a period of time due to the consistent overwatering, and its dry mass reaches 0.

```python
init_flood = model_plant_growth_v5(3,7,40, default_params, irrigate = True, time_ticks = 50)
late_flood = model_plant_growth_v5(3,7,20, default_params, irrigate = True, irrigate_qty = 40, time_ticks = 50)

draw_nice_graph(init_flood, late_flood)
```

In the two graphs above, we observe what happens when a plant starts out with too much water (left) and when a plant suddenly receives too much water (right). In the first case, the plant dies in the presence of too much water and its dry mass reaches 0. In the second case, the plant grows fine at the beginning until it gets watered with too much water, causing the plant to die.

## Improved Model Overview

Now that we have made the necessary improvements to our model to better align it with the scientific basis of photosynthetic processes within a plant, we have our final model, encapsulated in the function `model_plant_growth_v5()`.

This function takes in initial values for x, w, and v. It also takes in `parameters`, which will be further explained later. Optional arguments include:
- `time_ticks`: the number of time ticks to run the simulation for. Defaults to 300.
- `irrigate`: a boolean determining whether the plant receives irrigation. Defaults to False.
- `irrigate_threshold`: the $v(t)$ value below which the plant will receive irrigation. Defaults to 5.
- `irrigate_qty`: the quantity of water to be added to the soil when the plant is irrigated. Defaults to 10.
- `drown_threshold`: the $v(t)$ value above which the plant begins to die. Defaults to 30.

The input `parameters` is an iterable (list or tuple) containing 7 elements. The elements are, in order:
1. `k` :  A modifier on the mass produced by the plant per unit time per unit mass of water.
2. `g` :  A modifier to limit the rate of increase of $x(t)$, preventing $x'(t)$ from getting too large.
3. `m` :  A modifier on the rate of degradation of the plant per unit mass, affects x(t) multiplicatively.
4. `q` :  A modifier on the rate of water uptake by the plant from the soil.
5. `r` :  A modifier to limit the rate of increase of $w(t)$, preventing $w'(t)$ from getting too large.
6. `b` :  A modifier on the rate of water loss from the plant to non-photosynthetic processes, such as transpiration or respiration.
7. `y` :  A modifier on the rate of water loss from the soil to non-plant sources, such as evaporation or leakage into the ground.

# Possible Extensions 

It is generally scientifically accepted that [C3 plants, on average, require 660ml of water to generate 1g of biomass. C4 plants require 300ml of water](https://www.jstor.org/stable/3987797) to generate 1g of biomass on average. For simplicity, we will take an intermediate value of 500ml of water to produce 1g of biomass. 

On top of that, [20%-30% of water expended on crops are lost through evaporation from the soil](https://water.unl.edu/documents/Section%20I.pdf). For simplicity, we can take this value to be 25% in our modelling. This gives us an indication of what our approximate `y` value should be.

Furthermore, approximately [99% of water taken up by plants are lost through transpiration](https://www.cambridge.org/core/books/abs/plant-physiology/transpiration/51D6090425F93DFD3AD8D69D5335F2E1). Of the remaining 1%,  roughly 90% of it is retained within cells for metabolic processes and only about 10% is used for photosynthesis. Hence, 99.9% of water taken up by a plant is used on non-photosynthetic processes, and only 0.01% of the water taken up by a plant is used for photosynthesis. This strongly suggests that our value of `b` should be approximately 0.999. 

Taking the example of 500ml of water to produce 1g of biomass, we know that only $500 \times 0.75 = 375$ml of water is taken up by the plant, and of this 375ml, $375\times0.001 = 0.375$ml is available for photosynthesis.

Because the ratio of water in the soil to the water available for photosynthesis is 500ml : 0.375ml, approximately 1000 : 1, we can let our units for $x(t)$, $w(t)$, and $v(t)$ be in grams, millilitres, and litres respectively. 

With these changes, we could either manipulate our formulas to account for conversion between these units, or we could simply edit our `parameter` elements to account for these discrepancies. 

Given more time and perhaps mathematical modelling software that could explore the introduction of constants into our mathematical equations, this model could be edited such that the constant parameter values are more grounded in scientific reasoning.

However, at the present moment, our model default parameters would have to be adjusted on a case-by-case basis dependent on the values produced by agricultural data, so as to fit the model to a particular growing environment or a specific crop's growing properties. Once this model is established with the proper parameters for a specific agricultural purpose, it can be used to model the expected biomass return based on differing irrigation quantities and thresholds. This would help to improve the efficiency of irrigation strategies without having to attempt their implementation, which could be costly in terms of time and money.

Furthermore, our function `model_plant_growth_v5()` does not plot on its own and instead returns data that can be plotted, so this allows for the raw data produced from our calculations to be extracted from the model output. For instance, if we want to obtain the list of values of $x(t)$ from one set of calculations, we could run the code below:

```python
a = model_plant_growth_v5(3,7,20, default_params,irrigate = True)
timepoints = a[0]
x_data = a[1]
w_data = a[2]
v_data = a[3]
```
